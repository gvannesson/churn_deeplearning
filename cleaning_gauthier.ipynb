```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('churn.csv')
```

```python
colonnes= data.columns
```

```python
categorical_column = [
    'gender','SeniorCitizen','Partner','Dependents','PhoneService','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport',
    'StreamingTV','StreamingMovies','Contract','PaperlessBilling'
]
```

```python
for col in colonnes:
    if col in categorical_column:
        categories = data[col].value_counts().index
        counts = data[col].value_counts().values
        plt.bar(categories, counts, width=0.5)
        plt.title(col)
        plt.show()
```
